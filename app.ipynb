### Snake Recognition Task

This task involves working with an image dataset containing pictures of snakes, each labeled with its species. The assignment consists of two main goals:

1. **Species Classification:**  
   Identify the species of snakes in previously unseen images.

2. **Venomous vs. Non-venomous Classification:**  
   Determine whether a given snake is venomous or not.

In practice, this means you will implement:
- A **multiclass classifier** for species recognition.
- A **binary classifier** for venomous/non-venomous detection.  
Both tasks should be solved using deep learning models without manual intervention (i.e., no human visual assistance).
